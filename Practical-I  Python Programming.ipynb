```python
# d) Accept a number and print whether it is prime of not. Extend the programme to list down the prime number between 1 and 100 and compute mean of the same.

num=int(input("Enter Number:"))
if num==1:
    print(num,"is neither prime or not composite")
else:
    i=2
    while i < num:
        if num%i==0:
            print (num,"is not prime number")
            break
        i=i+1
    if i==num:
        print(num,"is a prime number")
        
# prime numbers between 1 to 100
Prime_Numbers=[]
for num in range(2,101):
    i=2
    while i < num:
        if num%i==0:
            break
        i=i+1
    if i==num:
        Prime_Numbers.append(num)
print("Prime numbers between 1 to 100 is",Prime_Numbers)

# mean of prime Numbers
s=float(0)
for j in range(len(Prime_Numbers)):
    s=s+Prime_Numbers[j]
mean=s/len(Prime_Numbers)
print("mean of Prime numbers between 1 to 100 is",mean)

```

```python
# e) Accept two related numeric data series and print mean and variance of each series, covariance and correlation between them.

X=[21,45,33,44]
Y=[45,65,78,67]
# mean function
def mean(z):
    n=len(z)
    s=0
    for i in range(n):
        s=s+z[i]
    return(s/n)
# variance function
def variance(z):
    n=len(z)
    s=0
    xm=mean(z)
    for i in range(n):
        s=s+z[i]*z[i]
    return(s/n-xm**2)
# covariance function
def covariance(x,y):
    n=len(x)
    s=0
    for i in range(n):
        s=s+x[i]*y[i]
    return(s/n-mean(x)*mean(y))

print("mean of X is",mean(X))
print("mean of Y is",mean(Y))
print("Variance of X is",variance(X))
print("Variance of Y is",variance(Y))
print("covariance of X and Y is",covariance(X,Y))
print("correlation of X and Y is",covariance(X,Y)/(variance(X)**(1/2))*(variance(Y)**(1/2)))
```

```python
# g) To generate random number from Binomial, Poisson, Hypergeometric, Negative Binomial, Multinomial distributions using uniform random numbers

# binomial random number
import numpy as np
m=10
p=input("Enter value of p:")
n=input("input value of n:")
BRN=[]
for j in range(int(m)):
    s=0
    for i in range(int(n)):
        u=np.random.rand(1)
        if u<=float(p):
            s=s+1
    BRN.append(s)
print(BRN)

```

```python
# poisson
n=12
lm=4
z=[]
e=np.exp(-lm)
for j in range(n): 
    t=0
    s=1
    while(e<s):
        u=np.random.rand(1)
        s=s*u
        t=t+1
    z.append(t)
print(z)
```

```python
# Negative binomial
n=15
k=10
p=0.78
y=[]
for j in range(n):
    s=0
    for i in range(k):
        u=np.random.rand(1)
        x=int(np.log(u)/np.log(1-p))
        s=s+x
    y.append(s)
print(y)
```

```python
# multinominal
n=20
R=15
p=[0.3,0.4,0.1,0.1,0.1]
k=len(p)
pc=[0]+list(np.cumsum(p))
y=np.zeros((12,5),dtype=int)
for l in range(R):
    z=np.zeros((5),dtype=int)
    for j in range(n):
        u=np.random.rand(1)
        for i in range(k):
            if u>pc[i] and u<=pc[i+1]:
                z[i]=z[i]+1
                break
    y[l: ]=z
print(y)
```

```python
# h) To find square root of any number using Newton Raphson method

n=100
x0=input("Initial guess:")
error=1
def f_x(x):
    return(float(x)**2-n)
def f_dash_x(x):
    return(2*float(x))
while(error > 0.00001):
    x1=float(x0)-(f_x(x0)/f_dash_x(x0))
    error=abs(x1-float(x0))
    x0=x1
print(x1)
```

```python
# d) To find mean. Variance, coefficient of Skewness and Kurtosis of Y

x=stats.lognorm.rvs(size=100,s=3,scale=np.exp(1))
mean=stats.lognorm.mean(s=3,scale=np.exp(1))
var=stats.lognorm.var(s=3,scale=np.exp(1))
kurtosis=stats.kurtosis(x)
skew=stats.skew(x)
print("Mean =",mean)
print("Var =",var)
print("Kurtosis =",kurtosis)
print("Skew =",skew)
```
